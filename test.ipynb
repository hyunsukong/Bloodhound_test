```python
def sort_particle_by_dist(coords, com):
    part_dist = np.linalg.norm(coords - com, None, 1)
    # Sort the distance array
    sort_idx = np.argsort(part_dist)
    sorted_part_dist = part_dist[sort_idx]
    return(sorted_part_dist, sort_idx)
```

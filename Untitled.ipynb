```python
from matplotlib import pyplot as plt
import pandas as pd
import importlib
import numpy as np
```

```python
host_main_branch_fname = "/scratch/05097/hk9457/FIREII/m12c_r7100/tree_processed_data/host_main_branch.csv"
infalling_subtree_fname = "/scratch/05097/hk9457/FIREII/m12c_r7100/bloodhound_subhalo_tracking/infalling_subhalo_result/m12c_r7100_subtree_infall_subhalo.hdf5"
infalling_tree_fname = "/scratch/05097/hk9457/FIREII/m12c_r7100/bloodhound_subhalo_tracking/infalling_subhalo_result/m12c_r7100_tree_infall_subhalo.hdf5"
```

```python
host_main_branch_df = pd.read_csv(host_main_branch_fname)
infalling_subtree_df = pd.read_hdf(infalling_subtree_fname)
infalling_tree_df = pd.read_hdf(infalling_tree_fname)
merged_df = pd.concat([infalling_subtree_df, infalling_tree_df])
infalling_only_df = merged_df.query("`infalling?`==1")
```

```python
infall_tids = np.unique(merged_df['tree.tid'].values)
infall_grouped_by_tid = merged_df.groupby(by='tree.tid')
infalling_only_tids = np.unique(infalling_only_df['tree.tid'].values)
infalling_only_grouped_by_tid = infalling_only_df.groupby(by='tree.tid')
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(host_main_branch_df['snapshot'].values,
        host_main_branch_df['radius'].values)
for i in range(len(infall_tids)):
    tid = infall_tids[i]
    tree = infall_grouped_by_tid.get_group(tid)
    infall_snap = tree['snapshot.infall'].values[0]
    infall_idx = np.where(tree['snapshot'].values == infall_snap)[0][0]
    infall_row = tree.iloc[infall_idx]
    ax.plot(infall_row['snapshot'], infall_row['dist'], 'o', c='red', rasterized=True, alpha=0.7)
print(len(infalling_only_tids))
counter = 0
for i in range(len(infalling_only_tids)):
    tid = infalling_only_tids[i]
    tree = infalling_only_grouped_by_tid.get_group(tid)
    infall_snap = tree['snapshot.infall'].values[0]
    infall_idx = np.where(tree['snapshot'].values == infall_snap)[0][0]
    infall_row = tree.iloc[infall_idx]
    if infall_row['ID.halo.infall'] < 0:
        counter += 1
    ax.plot(infall_row['snapshot'], infall_row['dist'], 'o', c='black', rasterized=True, alpha=0.7)
ax.set(yscale='log')
print(counter)
```
